**TOPIC : WHILE LOOP**

With the while loop we can execute a set of statements as long as a condition is true.

**SYNTAX:**
Syntax. An expression evaluated before each pass through the loop. If this condition evaluates to true, statement is executed. When condition evaluates to false, execution continues with the statement after the while loop.


*   while (condition)
*   statement item

1. Print all elements of a list using a while loop.

```python
my_list = [1,2,3,4,5]
i = 0
while i < len(my_list):
    print(my_list[i])
    i += 1
```

2. Print all even numbers in a list using a while loop.

```python
numbers = [10, 15, 22, 33, 40, 55]
i = 0
while i < len(numbers):
    if numbers[i] % 2 == 0:
        print(numbers[i])
    i += 1
```

3. Print list elements in reverse order using a while loop.



```python
my_list = [1, 2, 3, 4, 5]
i = len(my_list) - 1
while i >= 0:
    print(my_list[i])
    i -= 1
```

4. Calculate the sum of elements in a list.



```python
numbers = [5, 10, 15, 20]
i = 0
total = 0
while i < len(numbers):
    total += numbers[i]
    i += 1
print("Sum:", total)
```

5. Count how many odd numbers are in a list.



```python
numbers = [1, 4, 7, 8, 11, 14]
i = 0
odd_count = 0
while i < len(numbers):
    if numbers[i] % 2 != 0:
        odd_count += 1
    i += 1
print("Odd numbers count:", odd_count)
```

6. Find the maximum value in a list using a while loop.



```python
numbers = [23, 89, 15, 78, 99, 5]
i = 1
max_val = numbers[0]
while i < len(numbers):
    if numbers[i] > max_val:
        max_val = numbers[i]
    i += 1
print("Maximum:", max_val)
```

8. Print the square of each number in a list using while loop.

9. Print only the string elements from a mixed list.



```python
mixed_list = [10, "hello", 3.14, "world", True, "Python"]
i = 0
while i < len(mixed_list):
    if type(mixed_list[i]) == str:
        print(mixed_list[i])
    i += 1
```

10. Calculate the average of all numbers in a list.

```python
numbers = [10, 20, 30, 40, 50]
i = 0
total = 0
while i < len(numbers):
    total += numbers[i]
    i += 1
average = total / len(numbers)
print("Average:", average)
```

11. Count how many times a specific value appears in a list.



```python
my_list = [1, 3, 5, 3, 7, 3, 9]
target = 3
i = 0
count = 0

while i < len(my_list):
    if my_list[i] == target:
        count += 1
    i += 1

print(f"{target} appears {count} times.")
```

12. Create a new list that contains only positive numbers from the original list.



```python
original = [-5, 3, -2, 7, -1, 9]
positive = []
i = 0

while i < len(original):
    if original[i] > 0:
        positive.append(original[i])
    i += 1

print("Positive numbers:", positive)
```

13. Remove all odd numbers from a list using while loop.



```python
numbers = [1, 2, 3, 4, 5, 6]
i = 0

while i < len(numbers):
    if numbers[i] % 2 != 0:
        numbers.pop(i)
    else:
        i += 1

print("After removing odd numbers:", numbers)
```

14. Check if a specific element exists in the list using while loop (no in keyword).



```python
my_list = [10, 20, 30, 40]
target = 30
i = 0
found = False

while i < len(my_list):
    if my_list[i] == target:
        found = True
        break
    i += 1

if found:
    print(f"{target} exists in the list.")
else:
    print(f"{target} does not exist in the list.")
```

15. Merge two lists using while loop.



```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
merged = []
i = 0

while i < len(list1):
    merged.append(list1[i])
    i += 1

i = 0
while i < len(list2):
    merged.append(list2[i])
    i += 1

print("Merged list:", merged)
```

16. Separate even and odd numbers from a list into two new lists.



```python
numbers = [10, 15, 22, 33, 40, 55]
i = 0
even = []
odd = []

while i < len(numbers):
    if numbers[i] % 2 == 0:
        even.append(numbers[i])
    else:
        odd.append(numbers[i])
    i += 1

print("Even numbers:", even)
print("Odd numbers:", odd)
```

17. Find and print duplicate elements in a list using while loop.

```python
my_list = [2, 4, 2, 6, 4, 8, 10, 6]
duplicates = []
i = 0

while i < len(my_list):
    j = i + 1
    while j < len(my_list):
        if my_list[i] == my_list[j] and my_list[i] not in duplicates:
            duplicates.append(my_list[i])
        j += 1
    i += 1

print("Duplicate elements:", duplicates)
```
