**FUNCTIONS**

In Python, a function is defined using the def keyword, followed by the function name, parentheses containing any parameters, and a colon. The function body is indented and contains the code to be executed.

**Types of Functions**

1. Built-in Functions: These are pre-defined functions provided by a programming language or environment. Examples include print(), len() in Python, and printf() in C.

2. User-defined Functions: These are functions created by the programmer to perform specific tasks. They follow a defined structure with a name, parameters, and a body.

3. Anonymous Functions (Lambda Functions): These are functions without a name, often used for short, simple operations

4. Recursive Functions: These functions call themselves within their definition, useful for tasks that can be divided into similar subtasks.

```python
#Recursive function
def factorial(n):
    # Base case: if n is 0, return 1
    if n == 0:
        return 1
    # Recursive case: n! = n * (n-1)!
    else:
        return n * factorial(n - 1)

# Using the recursive function
result = factorial(7)
print(result)
```

```python
def factorial(n):
    # Base case: if n is 0, return 1
    if n == 0:
        return 1
    # Recursive case: n! = n * (n-1)!
    else:
        return n * factorial(n - 1)

# Get user input
number = int(input("Enter a number to calculate its factorial: "))

# Ensure the input is a non-negative integer
if number < 0:
    print("Please enter a non-negative integer.")
else:
    # Using the recursive function
    result = factorial(number)
    print(f"The factorial of {number} is {result}")
```

**Arguments in Functions**

*Arguments are values passed to a function when it is called. They can be categorized as follows:*

1. Positional Arguments: These are arguments that are passed to a function in a specific order.

2. Keyword Arguments: These are arguments passed to a function by explicitly naming each parameter and its corresponding value.

3. Default Arguments: These are arguments that assume a default value if no value is provided during the function call

4. Variable-length Arguments: *args (Non-keyword arguments): Allows a function to accept any number of positional arguments.

**kwargs (Keyword arguments): Allows a function to accept any number of keyword arguments

**Order of Positional and Keyword Arguments**

if there are both positional and keyword arguments, the positional ones must come first in the function call.

```python
def func(a, b, c, d):
  return a + b + c + d

func(4, 5, d = 1, c = 6)
```

**Otherwise it won’t work:**

```python
func(c = 6, d = 1, 4, 5)
```

By the way, keyword arguments may be in any order.



**Order of Default and Mandatory Arguments **

if there are default arguments, they must follow mandatory arguments.

```python
def func(a, b, c = 3, d = 5):
  return a + b + c + d
```

If we don’t pass the default arguments, their default values will be used:



```python
func(1, 2)
```

If we do pass the default arguments, their default values will be overwritten:

```python
func(1, 2, 4, 8)
```

If we need both positional and keyword arguments in a function, then first come all single positional arguments, then we can use the * notation to pack all the remaining positional arguments in a tuple, and last come the keyword arguments packed in a dictionary using the ** notation:

```python
def func(a, *b, **c):
  print(a, b, c)
```


2. Create a list of numbers. Write a function that finds and returns the maximum value in the list without using the built-in max() function.



```python
def find_max(lst):
    if not lst:
        return None
    max_val = lst[0]
    for num in lst:
        if num > max_val:
            max_val = num
    return max_val
print("2. Maximum Value:", find_max([3, 7, 2, 9, 5]))
```

7. Write a recursive function that calculates the Fibonacci sequence for a given term.



```python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)
print("7. Fibonacci number at position 6:", fibonacci(6))
```


9. Write a function that takes a list of numbers and returns a new list containing only the unique elements.

```python
def unique_elements(lst):
    unique = []
    for item in lst:
        if item not in unique:
            unique.append(item)
    return unique
print("9. Unique elements:", unique_elements([1, 2, 2, 3, 4, 4, 5]))
```

10. Create a Python function that accepts a string and counts the occurrences of each character. Return the result as a dictionary.



```python
def count_characters(s):
    char_count = {}
    for char in s:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    return char_count

print("10. Character Count:", count_characters("banana"))
```

11. Implement a function to calculate the average of a list of numbers without using the built-in sum() and len() functions.


```python
def calculate_average(lst):
    total = 0
    count = 0
    for num in lst:
        total += num
        count += 1
    if count == 0:
        return 0
    return total / count

print("11. Average:", calculate_average([10, 20, 30, 40]))
```


12. Write a function that checks if a given year is a leap year. A leap year is divisible by 4 but not divisible by 100 unless it is divisible by 400.


```python
def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

print("12. Is 2024 Leap Year?:", is_leap_year(2024))
print("12. Is 1900 Leap Year?:", is_leap_year(1900))
print("12. Is 2000 Leap Year?:", is_leap_year(2000))
```


14. Create a function that accepts a sentence and counts the number of vowels (a, e, i, o, u) in it.



```python
def count_vowels(sentence):
    vowels = 'aeiouAEIOU'
    count = 0
    for char in sentence:
        if char in vowels:
            count += 1
    return count

print("14. Number of vowels:", count_vowels("This is a sample sentence."))
```

15. Write a Python function that takes a number and checks whether it is a prime number.


```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print("15. Is 7 prime?:", is_prime(7))
print("15. Is 10 prime?:", is_prime(10))
```


16. Implement a function that reverses the order of words in a given sentence.

```python
def reverse_words(sentence):
    words = sentence.split()
    return ' '.join(reversed(words))

print("16. Reversed words:", reverse_words("Python is a powerful language"))
```

17. Create a function to find and return the second-largest number in a list of integers.


```python
def second_largest(numbers):
    if len(numbers) < 2:
        return None
    first = second = float('-inf')
    for num in numbers:
        if num > first:
            second = first
            first = num
        elif first > num > second:
            second = num
    return second if second != float('-inf') else None

print("17. Second largest number:", second_largest([10, 5, 20, 8, 20]))
print("17. Second largest number (unique list):", second_largest([3, 7, 1]))
print("17. Second largest number (too short):", second_largest([5]))
```


18. Write a function that takes a string and returns True if it is a palindrome (ignoring spaces and case), and False otherwise.

```python
def is_clean_palindrome(s):
    cleaned = ''.join(char.lower() for char in s if char != ' ')
    return cleaned == cleaned[::-1]

print("18. 'Race car' is palindrome?:", is_clean_palindrome("Race car"))
print("18. 'Hello World' is palindrome?:", is_clean_palindrome("Hello World"))
print("18. 'Madam In Eden I'm Adam' is palindrome?:", is_clean_palindrome("Madam In Eden I'm Adam"))
```
